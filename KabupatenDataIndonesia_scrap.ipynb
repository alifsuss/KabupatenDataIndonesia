## Import Libraries

```python
import pandas as pd
import numpy as np
# import tensorflow as tf
import matplotlib.pyplot as plt
from requests import get
from bs4 import BeautifulSoup as Soup
from bs4 import NavigableString, Tag
import time 
```

## Get the site necessity 

```python
url = get("https://id.wikipedia.org/wiki/Daftar_kabupaten_dan_kota_di_Indonesia")
```

```python
request = url.text
```

### Parse the whole site 

```python
total_soup = Soup(request,'html.parser')
```

### Experiment First

```python
#Province names
names = total_soup.find('div', {'class':'mw-parser-output'}).findAll('h3')
```

```python
#This is the province name
display(names[3].find('span', {'class': 'mw-headline'}).text)
print(len(names))
```

```python
# This to get all tables for each provinces
tables = total_soup.find('div', {'class':'mw-parser-output'}).findAll('table', {'class': 'wikitable sortable mw-collapsible'})
```

## Now to the Action 

```python
# Declare the number of pages that available, in this case its 165 max

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:78.0) Gecko/20100101 Firefox/78.0'
}

#declare the url and the page variable
url = get("https://id.wikipedia.org/wiki/Daftar_kabupaten_dan_kota_di_Indonesia", headers=headers)
    
    
#create empty lists 
provinsi = []
kabupaten = []
ibukota = []
hektar = []
penduduk = []

start = time.time()

for i in range(len(tables)):
    print("Accessing {} province".format(names[i].find('span', {'class': 'mw-headline'}).text))
    
    for j in range(1,len(tables[i].findAll('tr'))):
        print("Entering Kabupaten/Kota {}".format(tables[i].findAll('tr')[j].findAll('td')[1].text))
        print("-----------------------------------------")
        
        #Province
        provinsi.append(names[i].find('span', {'class': 'mw-headline'}).text)
        
        #kabupaten
        kabupaten.append(tables[i].findAll('tr')[j].findAll('td')[1].text)
        
        #ibukota
        ibukota.append(tables[i].findAll('tr')[j].findAll('td')[2].text)
        
        #hektar
        hektar.append(tables[i].findAll('tr')[j].findAll('td')[4].text.replace('.','').replace(',',''))
        
        #penduduk
        penduduk.append(tables[i].findAll('tr')[j].findAll('td')[5].text.replace('.',''))
        
print("time spent {} seconds".format(time.time() - start))
```
