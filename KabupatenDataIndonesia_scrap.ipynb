## Now to the Action 

```python


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:78.0) Gecko/20100101 Firefox/78.0'
}

#declare the url and the page variable
url = get("https://id.wikipedia.org/wiki/Daftar_kabupaten_dan_kota_di_Indonesia", headers=headers)
    
    
#create empty lists 
provinsi = []
kabupaten = []
ibukota = []
hektar = []
penduduk = []

start = time.time()

for i in range(len(tables)):
    print("Accessing {} province".format(names[i].find('span', {'class': 'mw-headline'}).text))
    
    for j in range(1,len(tables[i].findAll('tr'))): #its start with 1, because row 0 correspond to the header
        print("Entering Kabupaten/Kota {}".format(tables[i].findAll('tr')[j].findAll('td')[1].text))
        print("-----------------------------------------")
        
        #Province
        provinsi.append(names[i].find('span', {'class': 'mw-headline'}).text.replace('\n',''))
        
        #kabupaten
        kabupaten.append(tables[i].findAll('tr')[j].findAll('td')[1].text.replace('\n',''))
        
        #ibukota
        ibukota.append(tables[i].findAll('tr')[j].findAll('td')[2].text.replace('\n',''))
        
        #hektar
        hektar.append(tables[i].findAll('tr')[j].findAll('td')[4].text.replace('.','').replace(',','').rstrip())
        
        #penduduk
        penduduk.append(tables[i].findAll('tr')[j].findAll('td')[5].text.replace('.','').rstrip())
        
print("time spent {} seconds".format(time.time() - start))
```

```python
himpun =[provinsi, kabupaten, ibukota, hektar, penduduk]
df = pd.DataFrame(himpun).T
df.columns = ["provinsi", "kabupaten", "ibukota", "hektar", "penduduk"]

display(df)
```

```python
df[['hektar','penduduk']] = df[['hektar','penduduk']].astype(float)
```

```python
df.dtypes
```

```python
df['density'] = df['penduduk'] / (df['hektar'] / 100)
```

```python
df.sort_values(by='density', ascending=False).head(20)
```
